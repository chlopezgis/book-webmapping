### **Verificar que los `ubigeo` se correspondan con la capa de `distritos`**

Buscando todos donde al unir `barrios` y `distritos` el nombre del distrito sea **nulo**

```python
%%sql
WITH resumen as 
    (SELECT a.ubigeo, b.distrito, count(*) FROM cartografia.barrios a
     LEFT JOIN cartografia.distritos b ON a.ubigeo = b.ubigeo
     GROUP BY 1,2)
     SELECT * FROM resumen where distrito is null;
```

Podemos corregir el de ubigeo `comas`:

```python
%%sql
SELECT * FROM cartografia.distritos WHERE upper(distrito) = 'COMAS';
```

```python
%%sql
UPDATE cartografia.barrios SET ubigeo = '150110' 
WHERE upper(ubigeo) = 'COMAS';
```

Solo para los de **`ubigeo`** nulos le asignaremos ubigeo de forma espacial:

```python
%%sql
UPDATE cartografia.barrios a SET ubigeo = b.ubigeo
FROM cartografia.distritos b WHERE st_intersects(st_centroid(a.geom), b.geom)
and a.ubigeo is null;
```

```python
%%sql
WITH resumen as 
    (SELECT a.ubigeo, b.distrito, count(*) FROM cartografia.barrios a
     LEFT JOIN cartografia.distritos b ON a.ubigeo = b.ubigeo
     GROUP BY 1,2)
     SELECT * FROM resumen where distrito is null;
```

### **Concatenar `nom_nhbd`, `comite`, `etapa`, `sector`, `zona` y `grupo`**

```python
%%sql
SELECT
    CONCAT(
        CASE WHEN nom_nhbd is not null THEN nom_nhbd else '' end,
        CASE WHEN comite is not null THEN ' COMITE '||comite else '' end,
        CASE WHEN etapa is not null THEN ' ETAPA '||etapa else '' end,
        CASE WHEN sector is not null THEN ' SECTOR '||sector else '' end,
        CASE WHEN zona is not null THEN ' ZONA '||zona else '' end,
        CASE WHEN grupo is not null THEN ' GRUPO '||grupo else '' end
        ) label_nhbd,
    comite, etapa, sector, zona, grupo
FROM cartografia.barrios
WHERE etapa is not null and zona is not null
LIMIT 10;
```

```python
%%sql
ALTER TABLE cartografia.barrios ADD COLUMN label_nhbd TEXT;

UPDATE cartografia.barrios SET label_nhbd =
                    CONCAT(
                            CASE WHEN nom_nhbd is not null THEN nom_nhbd else '' end,
                            CASE WHEN comite is not null THEN ' COMITE '||comite else '' end,
                            CASE WHEN etapa is not null THEN ' ETAPA '||etapa else '' end,
                            CASE WHEN sector is not null THEN ' SECTOR '||sector else '' end,
                            CASE WHEN zona is not null THEN ' ZONA '||zona else '' end,
                            CASE WHEN grupo is not null THEN ' GRUPO '||grupo else '' end
                        );
```

### **Verificar el campo `mz_muni`**

**Limpiando espacios**

```python
%%sql
UPDATE cartografia.manzanas SET mz_muni = TRIM(mz_muni) WHERE mz_muni LIKE ' %';
```

**Espacios y vacios**

```python
%%sql
WITH mz_res AS (
    SELECT mz_muni, count(*) 
    FROM cartografia.manzanas
    GROUP BY 1 ORDER BY 1)
    SELECT * FROM mz_res WHERE mz_muni IN ('', ' ');
```

```python
%%sql
--Reemplazando las '' y ' '
UPDATE cartografia.manzanas SET mz_muni = NULL WHERE mz_muni IN ('',' ','<Null>','<Nulo>');

--Verificando
WITH mz_res AS (
    SELECT mz_muni, count(*) 
    FROM cartografia.manzanas
    GROUP BY 1 ORDER BY 1)
    SELECT * FROM mz_res WHERE mz_muni IN ('', ' ','<Null>','<Nulo>');
```

**Con etiqueta MZ**

```python
%%sql
WITH mz_res AS (
    SELECT mz_muni, count(*) 
    FROM cartografia.manzanas
    GROUP BY 1 ORDER BY 1)
    SELECT * FROM mz_res WHERE mz_muni LIKE 'MZ%';
```

```python
%%sql
--Reemplazando las MZ
UPDATE cartografia.manzanas SET mz_muni = replace(mz_muni, 'MZ-','') WHERE mz_muni LIKE 'MZ-%';
UPDATE cartografia.manzanas SET mz_muni = replace(mz_muni, 'MZ','') WHERE mz_muni LIKE 'MZ%';

--Verificando
WITH mz_res AS (
    SELECT mz_muni, count(*) 
    FROM cartografia.manzanas
    GROUP BY 1 ORDER BY 1)
    SELECT * FROM mz_res WHERE mz_muni LIKE 'MZ%';
```

**Otras depuraciones**

```python
%%sql
--Reemplazando los '0' y '00'
UPDATE cartografia.manzanas SET mz_muni = NULL WHERE mz_muni IN ('0','00');

--Eliminado los '0' a la izquierda
UPDATE cartografia.manzanas SET mz_muni = SUBSTRING(mz_muni,2,10) WHERE mz_muni LIKE '0%';
```

```python
%%sql
--Reemplazando los 'UVC' y 'UCV' en ATE ('150103')
UPDATE cartografia.manzanas SET mz_muni = replace(mz_muni,'UVC','UCV')
WHERE mz_muni LIKE 'UVC%' AND ubigeo = '150103';

--Reemplazando 'UCV'  en ATE por NULL
UPDATE cartografia.manzanas SET mz_muni = NULL 
WHERE mz_muni = 'UCV' AND ubigeo = '150103';

--Reemplazando otros valores po NULL
UPDATE cartografia.manzanas SET mz_muni = NULL 
WHERE mz_muni in ('CEMENTERIO','COMERCIO', 'EMB EE.UU','S/N','SUB');
```
