### **2.1. Capas "clientes**

````python
!ogrinfo gas.gpkg clientes -al -so| grep -i 'count'
````

````python
!ogrinfo gas.gpkg clientes -al -fid 1 -q -nomd
````

````python
!ogrinfo gas.gpkg clientes -al -fid 1 -q -nomd \
 -sql "select OBJECTID as objectid, Tipo_Docum as tipdoc, \
              NUMERO_DOCUMENTO as numdoc, NOMBRE as nombre, \
              Saldo_Créd as saldo_cred, Cta_Contr_ as cta_contr, \
              DISTRITO as distrito, ZONA_DE_VENTA as zona_venta, \
              Dirección AS direccion, NSE as nse, MALLA as malla, \
              CORD_X as lon_x, CORD_Y as lat_y, \
              MakePoint(CORD_X, CORD_Y)\
        from clientes" \
````

Para guardarlo en el mismo geopackage contener utilizar la opción **`-update`**

````python
!ogr2ogr -f GPKG -a_srs EPSG:4326 \
 gas.gpkg -update gas.gpkg \
 -nlt POINT \
 -nln clientes_dep \
 -dialect SQLite \
 -geomfield geom \
 -sql "select OBJECTID as objectid, Tipo_Docum as tipdoc, \
              NUMERO_DOCUMENTO as numdoc, NOMBRE as nombre, \
              Saldo_Créd as saldo_cred, Cta_Contr_ as cta_contr, \
              DISTRITO as distrito, ZONA_DE_VENTA as zona_venta, \
              Dirección AS direccion, NSE as nse, MALLA as malla, \
              CORD_X as lon_x, CORD_Y as lat_y, \
              MakePoint(CORD_X, CORD_Y)\
        from clientes" \
````

Visualizar la capa exportada

````python
!ogrinfo gas.gpkg clientes_dep -al -so | grep -i 'count'
````

````python
!ogrinfo gas.gpkg clientes_dep -al -fid 1 -nomd
````

Se valido tambien con QGIS y es lo mismo, vamos a eliminar la capa original porque ser buffers

````python
!ogrinfo gas.gpkg -sql "DROP TABLE clientes"
````

#### **Mallavial y Crucemallavial**

````python
!ogrinfo gas.gpkg mallavial -al -so | grep -i 'count'
````

````python
!ogrinfo gas.gpkg mallavial -al -so | grep -i 'EPSG'
````

````python
!ogrinfo gas.gpkg mallavial -al -fid 1 -nomd -q
````

````python
!ogrinfo gas.gpkg \
-sql  "SELECT a.OBJECTID as objectid, a.CODIGOSEGMENTOVIA as cod_via,\
           a.CODIGOVIA as cod_nomvia, b.TIPOVIA as tip_via,  b.NOMBREVIA as nom_via, \
           b.SINONIMO as toponimo, a.UBIGEODERECHA as ubigeo_der, \
           a.UBIGEOIZQUIERDA as ubigeo_izq, a.NUMEROCUADRA as cuadra, \
           a.NUMEROINICIODERECHA as der_ini, a.NUMEROFINDERECHA as der_fin, \
           a.NUMEROINICIOIZQUIERDA as izq_ini, a.NUMEROFINIZQUIERDA as izq_fin, \
           a.NUMEROCUADRADERECHA as cuadra_der, a.NUMEROCUADRAIZQUIERDA as cuadra_izq, \
           a.FECHACREACION as created_date, a.FECHAMODIFICACION as last_edited_date, a.geom \
       FROM mallavial as a \
       LEFT JOIN (SELECT * FROM (SELECT *, ROW_NUMBER() OVER (PARTITION BY CODIGOVIA ORDER BY CODIGOVIA) AS rn \
                                 FROM crucemallavial) WHERE rn = 1) as b \
       ON a.CODIGOVIA = b.CODIGOVIA" \
-al -so | grep -i 'count'
````

````python
!ogr2ogr -f GPKG -a_srs EPSG:32718 -t_srs EPSG:4326 \
 gas.gpkg -update gas.gpkg \
 -nln ejes_viales \
 -nlt LINESTRING \
 -sql  "SELECT a.OBJECTID as objectid, a.CODIGOSEGMENTOVIA as cod_via,\
           a.CODIGOVIA as cod_nomvia, b.TIPOVIA as tip_via,  b.NOMBREVIA as nom_via, \
           b.SINONIMO as toponimo, a.UBIGEODERECHA as ubigeo_der, \
           a.UBIGEOIZQUIERDA as ubigeo_izq, a.NUMEROCUADRA as cuadra, \
           a.NUMEROINICIODERECHA as der_ini, a.NUMEROFINDERECHA as der_fin, \
           a.NUMEROINICIOIZQUIERDA as izq_ini, a.NUMEROFINIZQUIERDA as izq_fin, \
           a.NUMEROCUADRADERECHA as cuadra_der, a.NUMEROCUADRAIZQUIERDA as cuadra_izq, \
           a.FECHACREACION as created_date, a.FECHAMODIFICACION as last_edited_date, a.geom \
       FROM mallavial as a \
       LEFT JOIN (SELECT * FROM (SELECT *, ROW_NUMBER() OVER (PARTITION BY CODIGOVIA ORDER BY CODIGOVIA) AS rn \
                                 FROM crucemallavial) WHERE rn = 1) as b \
       ON a.CODIGOVIA = b.CODIGOVIA"
````

````python
!ogrinfo gas.gpkg \
-sql "SELECT a.OBJECTID as objectid, a.CODIGODISTRITO as ubigeo, \
              a.CODIGOPREDIO as cod_predio, \
              a.CODIGOSEGMENTOVIA as cod_via, a.CODIGOMANZANA as cod_mzna, \
              a.CODIGOPUERTA as cod_puerta, b.tip_via, b.nom_via, \
              a.NUMEROPUERTA as puerta_muni, c.mz_muni, a.NUMEROLOTE as lote_muni, a.geom \
        FROM puerta a \
        LEFT JOIN (SELECT * FROM (SELECT *, ROW_NUMBER() OVER (PARTITION BY cod_via ORDER BY cod_via) AS rn \
                                 FROM ejes_viales) WHERE rn = 1) b on a.CODIGOSEGMENTOVIA = b.cod_via \
        LEFT JOIN manzanas_dep c on a.CODIGOMANZANA = c.cod_mzna" \
-al -fid 221 -nomd -q
````
